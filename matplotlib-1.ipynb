```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(300, 2)
y = np.random.normal(loc=20.5, scale=1, size=(300, 2))

plt.title('test sample')
plt.xlabel('random num')
plt.ylabel('normal dist')

plt.plot(x, y, ls='none', marker='^')
plt.grid()
plt.show()
```

### Scatter plot

A scatter plot is a type of data visualization used to display the relationship between two continuous variables. It allows us to observe patterns, trends, and the strength of relationships between the variables.

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.random.normal(loc=20.5, scale=1, size=(300, 2))
bd = np.random.binomial(n=100, p=0.5, size=(300, 2))

ld = np.random.logistic(loc=2.25, scale=3.5, size=(300, 2))

# normal and bionomial distribution
plt.scatter(nd, bd, color='#88c999')

# normal and logistic distribution
plt.scatter(nd, ld)
plt.grid()
plt.show()
```
