```python
def print_hello():
    for i in range(20):
        if i == 10:
            time.sleep(2)
            print('................')
        print('__hello__')

def print_numbers(num):
    for i in range(num+1):
        print(i)


thread1 = threading.Thread(target=print_hello, args=())
thread2 = threading.Thread(target=print_numbers, args=(10))
print('__from the main thread__')

if __name__ == '__main__':
    # starting 2 threads
    thread1.start()
    thread2.start()
    thread1.join()
    thread2.join()
    print('This is also a main thread again!')
    print('Threads 1 & 2 has finished executing')
```

#### Without Multi-Threading

```python
import cv2
import time

delay = 0.03
vcap = cv2.VideoCapture(2)

if vcap.isOpened() is False:
    print('[Exiting]: Error while accessing the webcam.')
    exit(0)

fps_input_stream = int(vcap.get(5))
print(f"FPS of webcap hardware/input stream: {fps_input_stream}")

# grabbed, frame = vcap.read()

num_frames_processed = 0
st = time.time()

while True:
    grabbed, frame = vcap.read()
    if grabbed is False:
        print('[Exiting] No more frames to read.')
        break

    time.sleep(delay)

    new_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    num_frames_processed += 1

    cv2.imshow('frame', new_frame)
    if cv2.waitKey(1) and 0xFF == ord('q'):
        break
        exit(0)


vcap.release()
cv2.destroyAllWindows()

et = time.time()
elapsed = et - st
fps = num_frames_processed / elapsed

print(f"FPS: {fps}, Elapsed time: {elapsed}, Frame processed: {num_frames_processed}")


```

#### With Multi-Threading

```python
import cv2 
import time 
from threading import Thread 
import os
# import keyboard


class WebcamStream :
    def __init__(self, stream_id=0):
        self.stream_id = stream_id
        
        # opening video capture stream 
        self.vcap = cv2.VideoCapture(self.stream_id)
        if self.vcap.isOpened() is False :
            print("[Exiting]: Error accessing webcam stream.")
            exit(0)
        
        fps_input_stream = int(self.vcap.get(5))
        print('FPS of webcap hardware: ',fps_input_stream)

        self.grabbed, self.frame = self.vcap.read()
        if self.grabbed is False:
            print('[Exiting] No more frames to read.')
            # break
            exit(0)

        self.stopped = True

        self.thread = Thread(target=self.update, args=())
        # daemon threads keep running in the background while the program is executing 
        self.thread.daemon = True
    
    def start(self):
        self.stopped = False
        self.thread.start()

    def update(self):
        while True:
            if self.stopped is True:
                break
            self.grabbed, self.frame = self.vcap.read()
            if self.grabbed is False:
                print('[Exiting] No more frames to read')
                self.stopped = True
                break
        self.vcap.release()

    # method for returning latest read frame
    def read(self):
        return self.frame

    # method called to stop reading frames
    def stop(self):
        self.stopped = True




if __name__ == '__main__':

    os.environ["XDG_SESSION_TYPE"] = "xcb"

    webcap_stream = WebcamStream(stream_id=1)
    webcap_stream.start()

    num_frames_processed = 0
    start = time.time()

    while True:
        if webcap_stream.stopped is True:
            break
        else:
            frame = webcap_stream.read()
        
        delay = 0.03
        time.sleep(delay)
        num_frames_processed += 1

        new_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) 
        _new_frame = cv2.putText(img = new_frame,
                                text = "Good Morning",
                                org = (100, 100),
                                fontFace = cv2.FONT_HERSHEY_DUPLEX,
                                fontScale = 3.0,
                                color = (125, 246, 55),
                                thickness = 3
                                )

        cv2.imshow('frame', _new_frame)

        if cv2.waitKey(1) and 0xFFF == ord('q'):
            print('hellooo________')
            break

    end = time.time()
    webcap_stream.stop()

    elapsed = end - start
    fps = num_frames_processed / elapsed
    print(f'FPS: {fps}\nFrames processed: {num_frames_processed}\nElapsed time: {elapsed}')


print('hello')
```
