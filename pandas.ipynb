```python
import pandas as pd
import numpy as np

x = np.random.rand(100, 2)
y = np.random.normal(loc=1, scale=1, size=(100))

x1 = [ i[0] for i in x[0:, 0:1] ]
x2 = [ i[0] for i in x[0:, 1:] ]

data = {
    'x1': x1,
    'x2': x2,
    'y': y
}

df = pd.DataFrame(data)
# df
# creating files
targetFile = open('testdata.csv', 'a')
df.to_csv(targetFile, sep=',', index=False, encoding='utf-8')
```
